# **SeoulBike Data Analysis Project.**

* **Objective**

Analyze the factors affecting bike rentals in Seoul.

Understand patterns in bike usage over time and seasons.

Visualize the impact of weather and time on rental demand.

Provide insights to optimize bike rental operations.

```python
import pandas as pd     # Data Manipulation.
import numpy as np      # Numerical Computation
import seaborn as sns   # Statistical Visualization
import matplotlib.pyplot as plt # Data Plotting
```

```python
df=pd.read_csv("/SeoulBikeData (1).csv",encoding='cp1252')
```

```python
df.head()
```

* **Questionary**

   1.How does bike rental demand vary by time of day?

  2.What weather factors (temperature, humidity, etc.) affect rentals?

  3.Are there seasonal patterns in bike rentals?

  4.Do holidays impact bike rental demand?

  5.Can we predict rental demand based on weather and time?


```python
# Show the distribution of bike rentals by using Histogram.
sns.histplot(df['Rented Bike Count'], bins=30, kde=True)
plt.title('Distribution of Rented Bikes')
plt.show()
```

```python
# Show Bike Rentals Over Different Hours by using Line Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Hour', y='Rented Bike Count', data=df, ci=None)
plt.title('Bike Rentals by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rented Bikes')
plt.grid()
plt.show()
```

```python
#Show Bike Rentals by Season.
plt.figure(figsize=(8, 4))
sns.boxplot(x='Seasons', y='Rented Bike Count', data=df)
plt.title('Bike Rentals by Season')
plt.show()
```

```python
# Show impact of temperature on  bike Rentals - by using scatter plot..
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Temperature(째C)', y='Rented Bike Count', data=df)
plt.title('Temperature vs. Bike Rentals')
plt.show()
```

```python
#show Correlation Heatmap
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=np.number)    # Extracts only numerical columns from df.
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

* **Key Insights**

Bike rentals peak in the morning (7-9 AM) and evening (5-8 PM).

Higher temperatures lead to increased rentals, but extreme heat may reduce demand.

Rainfall and snowfall negatively impact bike rentals.

Winter has lower rentals, while summer has the highest.

Holidays see fewer rentals compared to working days.
